```python
import json
target_dir = "/home/mert/ps2/openevolve/examples/metagpt_evolution/"
combined_scores = []
total_failures = []
for i in range(10, 101, 10):
    with open(f"{target_dir}/openevolve_output/checkpoints/checkpoint_{i}/best_program_info.json", "r") as f:
        log = json.load(f)
    combined_scores.append(log['metrics']['combined_score'])
    total_failures.append(log['metrics']['avg_failures_per_task'])

import matplotlib.pyplot as plt
plt.plot(combined_scores)
plt.plot(total_failures)
plt.legend(['combined_score', 'avg_failures_per_task'])
plt.show()
```
